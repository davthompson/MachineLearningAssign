```python
import pandas as pd
import numpy as np

# Import the real data
df1 = pd.read_csv("2018_Financial_Data.csv", header = 0)


# Create the ratios dataframe. This dataframe will have 6 features that we need to do our clustering.
# First with the stock ticker
ratios = pd.DataFrame(df1, columns = ['Symbol']) 

# Calculate earnings per share ratio
ratios['EarningsPerShare'] = (df1['Net Income'] - df1['Preferred Dividends']) / df1['Weighted Average Shs Out']

# Get the P/E ratio directly from data
ratios['PriceEarningsRatio'] = df1['PE ratio']

# Calculate debt to equity ratio
ratios['DebtEquityRatio'] = (df1['Short-term debt'] + df1['Long-term debt']) / df1['Total shareholders equity']

# Get the return on equity ratio
ratios['ReturnOnEquity'] = df1['returnOnEquity']

# Get the quick ratio
ratios['QuickRatio'] = df1['quickRatio']

# Calculate the working capital ratio
ratios['WorkingCapitalRatio'] = df1['Total current assets'] / df1['Total current liabilities']


# Get rid of empty values from the date given
nan_value = float("NaN")
ratios.replace("", nan_value, inplace=True)
#ratios.replace(0, nan_value, inplace=True) #get rid of zero values?
ratios.dropna(subset = ['Symbol', 'EarningsPerShare', 'PriceEarningsRatio', 'DebtEquityRatio', 'ReturnOnEquity',
                       'QuickRatio', 'WorkingCapitalRatio'], inplace=True)
ratios = ratios.round(4)


#print(np.count_nonzero(ratios, axis = 0)) #Show number of non-zero values in the dataframe columns

# Show the date frame
#ratios = ratios.set_index('Symbol')
display(ratios)

# Put all these in a new CSV file
ratios.to_csv("2018_New_Financial_Data.csv", index = None)
```
