```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() #this is used for plot styling
from sklearn.cluster import KMeans

df1 = pd.read_csv("2018_New_Financial_Data.csv")
```

```python
financial_ratios_df = df1

# Get rid of empty values from the date given
nan_value = float("NaN")
financial_ratios_df.replace("", nan_value, inplace=True)
financial_ratios_df.replace(0, nan_value, inplace=True) #get rid of zero values?
financial_ratios_df.dropna(subset = ['Symbol', 'EarningsPerShare', 'PriceEarningsRatio', 'DebtEquityRatio', 'ReturnOnEquity',
                       'QuickRatio', 'WorkingCapitalRatio'], inplace=True)
financial_ratios_df = financial_ratios_df.round(4)
financial_ratios_df.to_csv("Target_Data.csv")

del financial_ratios_df['Symbol']
display(financial_ratios_df)
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(financial_ratios_df)
y_kmeans = kmeans.predict(financial_ratios_df)
```
